13. **What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

Ans:

*    The Central Limit Theorem (CLT) is one of the most important principles in statistics. It describes how the distribution of sample means becomes approximately normal (bell-shaped) as the sample size increases, regardless of the original distribution of the population, provided certain conditions are met.
- It is critical in statistics because:
  - The CLT allows statisticians to make inferences about population parameters using sample data, even when the population distribution is unknown.
  - Many statistical tests (e.g., z-tests, t-tests) and methods for constructing confidence intervals rely on the assumption of normality in the sampling distribution, which is guaranteed by the CLT for large samples.
  - It ensures that complex or non-normal distributions of population data can be analyzed using the simpler tools of the normal distribution.

20. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy.stats import norm

data = [10, 12, 23, 18, 20, 25, 19, 21, 17, 15]
confidence=0.95
data = np.array(data)

mean = np.mean(data)
std_dev = np.std(data, ddof=1)

n = len(data)
z_value = norm.ppf(1 - (1 - confidence) / 2)
margin_of_error = z_value * (std_dev / np.sqrt(n))
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

print(f"Confidence Interval ({confidence * 100}%): [{lower_bound:.4f}, {upper_bound:.4f}]")
```
